```python
import pandas as pd
import matplotlib.pyplot as plt

from utils import load_train_test
```

```python
train_filename = "application_train_numerical.csv"
path_to_data = "../data/"
```

### load train numerical

```python
df_application_train_numerical = load_train_test(
    filename=train_filename,
    path_to_train_folder=path_to_data,
)
```

```python
# check if greater value of feature leads to better target 0

greater_feature_is_better = []
greater_feature_is_worse = []


for column in df_application_train_numerical.columns[1:]:  # exclude target
    
    numerical_feature = df_application_train_numerical[column]
    numerical_feature_median = numerical_feature.median()
    
    numerical_feature_to_binary = (numerical_feature > numerical_feature_median)
    numerical_feature_to_binary.replace(
        to_replace={
            True: "greater_median",
            False: "less_median",
        },
        inplace=True,
    )
    
    target_crosstab = pd.crosstab(
        index=numerical_feature_to_binary,
        columns=df_application_train_numerical['TARGET'],
        margins=True,
    )
    display(target_crosstab)
    
    greater_median_1_target_rate = target_crosstab[1]["greater_median"] / target_crosstab.loc["greater_median", [0, 1]].sum()
    less_median_1_target_rate = target_crosstab[1]["less_median"] / target_crosstab.loc["less_median", [0, 1]].sum()
    
    print(f"bad target rate for 'greater median' clients: {round(greater_median_1_target_rate, 3)}")
    print(f"bad target rate for 'less median' clients: {round(less_median_1_target_rate, 3)}")
    
    if greater_median_1_target_rate > less_median_1_target_rate:
        print("greater values of feature is worse")
        greater_feature_is_worse.append(column)
    else:
        print("greater values of feature is better")
        greater_feature_is_better.append(column)
```

```python
greater_feature_is_better
```

```python
greater_feature_is_worse
```

### ranking

```python
greater_feature_is_better_rank = df_application_train_numerical[greater_feature_is_better].rank(
    ascending=False, na_option='bottom'
)
```

```python
greater_feature_is_worse_rank = df_application_train_numerical[greater_feature_is_worse].rank(
    ascending=True, na_option='bottom'
)
```

```python
ranked_data = pd.concat(
    [greater_feature_is_better_rank, greater_feature_is_worse_rank],
    axis=1,
)
```

```python
ranked_data.head()
```

```python
ranked_data_mean = pd.DataFrame(ranked_data.mean(axis=1), columns=['rank'])
ranked_data_mean['TARGET'] = df_application_train_numerical['TARGET']
```

```python
ranked_data_mean.head()
```

### plot results

```python
plt.figure(figsize=(20, 10))

rank_target_0 = ranked_data_mean[ranked_data_mean['TARGET'] == 0]['rank']  # GOOD
rank_target_1 = ranked_data_mean[ranked_data_mean['TARGET'] == 1]['rank']  # BAD

plt.scatter(rank_target_0, len(rank_target_0) * [0], color='lightgreen')
plt.scatter(rank_target_1, len(rank_target_1) * [0], color='red')
```
