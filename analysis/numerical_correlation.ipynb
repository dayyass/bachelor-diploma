```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from typing import Tuple

from stats_test import pearson_spearman_significance_test, kendall_significance_test
from utils import load_train_test
```

```python
train_filename = "application_train_numerical.csv"
path_to_data = "../data/"
```

### load train numerical

```python
df_application_train_numerical = load_train_test(
    filename=train_filename,
    path_to_train_folder=path_to_data,
)
```

### correlation

```python
corr_matrix = df_application_train_numerical.corr(method='pearson')
corr_matrix
```

```python
spearman_corr_matrix = df_application_train_numerical.corr(method='spearman')
spearman_corr_matrix
```

```python
kendall_corr_matrix = df_application_train_numerical.corr(method='kendall')
kendall_corr_matrix
```

### stats test

```python
n_columns = df_application_train_numerical.shape[1]
n_columns
```

```python
corr_matrix_p_value = np.zeros((n_columns, n_columns))
spearman_corr_matrix_p_value = np.zeros((n_columns, n_columns))
kendall_corr_matrix_p_value = np.zeros((n_columns, n_columns))
```

```python
for i in tqdm(range(n_columns)):
    for j in range(n_columns):
        
        col_1 = df_application_train_numerical.iloc[:, i]
        col_2 = df_application_train_numerical.iloc[:, j]
        
        n_samples = sum(~(col_1.isna() | col_2.isna()))
        
        corr = corr_matrix.iloc[i, j]
        spearman_corr = spearman_corr_matrix.iloc[i, j]
        kendall_corr = kendall_corr_matrix.iloc[i, j]
        
        corr_matrix_p_value[i, j] = pearson_spearman_significance_test(
            corr_coef=corr,
            n_samples=n_samples,
            alternative='two-sided',
        )[1]
        spearman_corr_matrix_p_value[i, j] = pearson_spearman_significance_test(
            corr_coef=spearman_corr,
            n_samples=n_samples,
            alternative='two-sided',
        )[1]
        kendall_corr_matrix_p_value[i, j] = kendall_significance_test(
            kendall_coef=kendall_corr,
            n_samples=n_samples,
            alternative='two-sided',
        )[1]
```

```python
corr_matrix_p_value = pd.DataFrame(
    corr_matrix_p_value,
    index=df_application_train_numerical.columns,
    columns=df_application_train_numerical.columns,
)
corr_matrix_p_value
```

```python
spearman_corr_matrix_p_value = pd.DataFrame(
    spearman_corr_matrix_p_value,
    index=df_application_train_numerical.columns,
    columns=df_application_train_numerical.columns,
)
spearman_corr_matrix_p_value
```

```python
kendall_corr_matrix_p_value = pd.DataFrame(
    kendall_corr_matrix_p_value,
    index=df_application_train_numerical.columns,
    columns=df_application_train_numerical.columns,
)
kendall_corr_matrix_p_value
```

### target unnormalized correlation

```python
for column in df_application_train_numerical.columns[1:]:
    target_filter = df_application_train_numerical['TARGET'] == 1
    
    feature_mean_target_1 = df_application_train_numerical[column][target_filter].mean()
    feature_mean_target_0 = df_application_train_numerical[column][~target_filter].mean()
    
    target_unnormalized_correlation = feature_mean_target_1 - feature_mean_target_0
    
    print(f"{column} target corr: {round(target_unnormalized_correlation, 3)}")
```
