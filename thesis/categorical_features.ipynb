```python
import pandas as pd
from IPython.display import display
from scipy.stats import chi2_contingency
from sklearn.metrics import matthews_corrcoef
from sklearn.preprocessing import LabelEncoder

from utils import load_train_test
```

```python
train_filename = "application_train_categorical.csv"
path_to_data = "../data/"
```

### load train categorical

```python
df_application_train_categorical = load_train_test(
    filename=train_filename,
    path_to_train_folder=path_to_data,
)
```

### contingency matrix

```python
# chi-square test of independence of variables in a contingency matrix
for column in df_application_train_categorical.columns[1:]:  # exclude target
    
    feature = df_application_train_categorical[column]
    target = df_application_train_categorical['TARGET']
    
    target_crosstab = pd.crosstab(
        index=feature,
        columns=target,
        margins=True,
    )
    display(target_crosstab)
    
    _, p_value, _, _ = chi2_contingency(target_crosstab)
    print(f"p-value: {round(p_value, 3)}\n")
```

```python
# chi-square test significance matthews_corrcoef
for column in df_application_train_categorical.columns[1:]:  # exclude target
    
    n_samples = sum(~df_application_train_categorical[column].isna())
    target = df_application_train_categorical['TARGET']
    
    if pd.api.types.is_string_dtype(df_application_train_categorical[column]):
        feature = LabelEncoder().fit_transform(df_application_train_categorical[column])
    else:
        feature = df_application_train_categorical[column]
    
    matthews_coef = matthews_corrcoef(
        y_pred=feature,
        y_true=target,
    )
    
    print(f"{column} matthews corr: {round(matthews_coef, 3)}")
```
